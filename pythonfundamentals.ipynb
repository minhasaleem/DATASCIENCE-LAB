ARITHMETIC OPERATORS

```python
a=5
b=2
sum=a+b
print("sum:",sum)
diff=a+b
print("diff:",diff)
product=a*b
print("product:",product)
mod=a%b
print("mod:",mod)
pow=a**b
print("pow:",pow)
```

list

```python
numbers = [10, 20, 30, 40]

print(numbers[1:3])
print(numbers[1:])
print(numbers[:3])

numbers.append(50)
print(numbers)

numbers.remove(20)
print(numbers)
```

tuple

```python
t = (1, 2, 3, 4, 5, 6)

print(t[2:4])
print(t[:3])
print(t[3:])
print(t[::-1])

print(len(t))

print(4 in t)
```

set

```python
s = {15, 25, 35, 65, 75}
print(s)

s.add(85)
print(s)

s.remove(35)
print(s)

print(25 in s)

print(len(s))
```

Dictionary

```python
student = {
    "name": "Minha",
    "age": 23
}

print(student)
print(student["age"])

student["age"] = 24
print(student)
```

string

```python
m = "minha"

print(m[1])

print(m[2:4])

b = "saleem"
print(m + " " + b)

print(len(m))

print(m.upper())
```

comparison operators

```python
a = 4
b = 6

print(a == b)
print(a != b)
print(a < b)
print(a > b)
print(a <= b)
print(a >= b)
```

if elif else

```python
marks = 70

if marks >= 90:
    print("Grade: A")
elif marks >= 80:
    print("Grade: B")
elif marks >= 70:
    print("Grade: C")
elif marks >= 60:
    print("Grade: D")
else:
    print("Grade: F")
```

nested if

```python
username = "admin"
password = "password123"

if username == "admin" :
    if password == "password123":
        print("Login successful")

```
