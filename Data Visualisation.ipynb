```python
import matplotlib.pyplot as plt
import numpy as np
x=np.linspace(0,5,11)
y=x**2
x
```

```python
y
```

```python
plt.plot(x,y,'b')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('X-Y graph')
plt.show()
```

```python
plt.subplot(1,3,2)
plt.plot(x,y,'b--')
plt.subplot(1,3,1)
plt.plot(x,y,'r*-')
plt.subplot(1,3,3)
plt.plot(x,y,'g^')
```

```python
fig = plt.figure()
axes = fig.add_axes([0.5,0.9,0.99,0.89])
axes.plot(x,y,'b')
axes.set_xlabel('X Axis')
axes.set_ylabel('Y Axis')
axes.set_title('X-Y graph')
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 100)
y = x**2

fig = plt.figure(figsize=(8,6))

# Main graph
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes1.plot(x, y, 'b')
axes1.set_xlabel('X Axis')
axes1.set_ylabel('Y Axis')
axes1.set_title('X-Y Graph')

# Small graph on the left side
axes2 = fig.add_axes([0.18, 0.55, 0.25, 0.25])
axes2.plot(y, x, 'r')
axes2.set_xlabel('X Axis')
axes2.set_ylabel('Y Axis')
axes2.set_title('X-Y Graph')

plt.show()
```

```python
fig,axes = plt.subplots()
axes.plot(x,y,'b')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('x-y graph');
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)
plt.Text(0.5, 1.0, 'X-Y Graph')
```

```python
for ax in axes:
    ax.plot(x,y,'b')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('x-y graph')
    fig.tight_layout()
fig
```

```python
fig = plt.figure(figsize = (8,4),dpi = 100 )
```

```python
fig,axes = plt.subplots(figsize=(8,3)) 
axes.plot(x,y,'b') 
axes.set_xlabel('x') 
axes.set_ylabel('y') 
axes.set_title('x-y graph') 
plt.Text(0.5, 1.0, 'x-y graph')
```

```python
fig.savefig('filename.png',dpi =200) 
#legend 
fig = plt.figure() 
ax = fig.add_axes([0,0,1,1]) 
ax.plot(x,x**2,label='x**2') 
ax.plot(x,x**3,label = 'x**3') 
ax.legend(loc = 2)
```

```python
ax.legend(loc = 1) #upper right 
ax.legend(loc =2)#upper left 
ax.legend(loc = 3) #lower left 
ax.legend(loc = 4) #lower right 
ax.legend(loc = 0 )#let matplotlib decide the optimal location
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define x values
x = np.arange(1,6)

# First line: y = x + 3
plt.plot(
    x, x + 3,
    color='b',
    linewidth=3,
    linestyle='--',
    marker='o',
    markersize=12,
    markeredgecolor='green',
    markeredgewidth=4,
    markerfacecolor='red'
)

# Second line: y = x + 4
plt.plot(
    x, x + 4,
    color='r',
    linewidth=4,
    linestyle='--',
    marker='o',
    markersize=10,
    markeredgecolor='blue',
    markeredgewidth=5,
    markerfacecolor='green'
)

# Display the plot
plt.show()
```
