```python
import matplotlib.pyplot as plt
import numpy as np
x=np.linspace(0,5,11)
y=x**2
x
```

```python
y
```

```python
plt.plot(x,y,'b')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('X-Y graph')
plt.show()
```

```python
plt.subplot(1,3,2)
plt.plot(x,y,'b--')
plt.subplot(1,3,1)
plt.plot(x,y,'r*-')
plt.subplot(1,3,3)
plt.plot(x,y,'g^')
```

```python
fig = plt.figure()
axes = fig.add_axes([0.5,0.9,0.99,0.89])
axes.plot(x,y,'b')
axes.set_xlabel('X Axis')
axes.set_ylabel('Y Axis')
axes.set_title('X-Y graph')
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 100)
y = x**2

fig = plt.figure(figsize=(8,6))

# Main graph
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes1.plot(x, y, 'b')
axes1.set_xlabel('X Axis')
axes1.set_ylabel('Y Axis')
axes1.set_title('X-Y Graph')

# Small graph on the left side
axes2 = fig.add_axes([0.18, 0.55, 0.25, 0.25])
axes2.plot(y, x, 'r')
axes2.set_xlabel('X Axis')
axes2.set_ylabel('Y Axis')
axes2.set_title('X-Y Graph')

plt.show()
```

```python
fig,axes = plt.subplots()
axes.plot(x,y,'b')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('x-y graph');
```
